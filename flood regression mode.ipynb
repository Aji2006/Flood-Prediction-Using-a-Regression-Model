## Transform the Outliers

```python
Q1 = train_df.quantile(0.25)
Q3 = train_df.quantile(0.75)
IQR = Q3 - Q1

condition = ~((train_df < (Q1 - 1.5 * IQR)) | (train_df > (Q3 + 1.5 * IQR))).any(axis=1)
df = train_df.loc[condition, train_df.columns]
```

## Standard Scaler

```python
numeric_features = df.select_dtypes(include=['number']).columns
numeric_features
 
scaler = StandardScaler()
df[numeric_features] = scaler.fit_transform(df[numeric_features])
```

## Dataset Splitting

```python
X = df.drop(columns=['FloodProbability'])
y = df['FloodProbability']
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
print(f"Total amount of data        : {len(X)}")
print(f"Total number of train data  : {len(x_train)}")
print(f"Total number of test data   : {len(x_test)}")
```

## LARS

```python
lars = linear_model.Lars(n_nonzero_coefs=1).fit(x_train, y_train)
pred_lars = lars.predict(x_test)
```

```python
mae_lars    = mean_absolute_error(y_test, pred_lars)
mse_lars    = mean_squared_error(y_test, pred_lars)
r2_lars     = r2_score(y_test, pred_lars)
rmse_lars   = np.sqrt(mse_lars)
 
print(f"MAE     : {mae_lars}")
print(f"MSE     : {mse_lars}")
print(f"R²      : {r2_lars}")
print(f"RMSE    : {rmse_lars}")
```

```python
data_eval = {
    "MAE" : [mae_lars],
    "MSE" : [mse_lars],
    "R2"  : [r2_lars],
    "RMSE" : [rmse_lars] 
}
df_eval = pd.DataFrame(data_eval, index=["Lars"])
df_eval
```

## LR (Linear Regression)

```python

LR = LinearRegression().fit(x_train, y_train)
pred_LR = LR.predict(x_test)

mae_LR   = mean_absolute_error(y_test, pred_LR)
mse_LR   = mean_squared_error(y_test, pred_LR)
r2_LR    = r2_score(y_test, pred_LR)
rmse_LR  = np.sqrt(mse_LR)

print(f"MAE     : {mae_LR}")
print(f"MSE     : {mse_LR}")
print(f"R²      : {r2_LR}")
print(f"RMSE    : {rmse_LR}")
```

```python
df_eval.loc['Linear Regression'] = [mae_LR, mse_LR, r2_LR, rmse_LR]
df_eval
```

## GradientBoostingRegressor

```python
GBR = GradientBoostingRegressor(random_state=184)
GBR.fit(x_train, y_train)

pred_GBR = GBR.predict(x_test)

mae_GBR     = mean_absolute_error   (y_test, pred_GBR)
mse_GBR     = mean_squared_error    (y_test, pred_GBR)
r2_GBR      = r2_score              (y_test, pred_GBR)
rmse_GBR    = np.sqrt               (mse_GBR)

print(f"MAE     : {mae_GBR}")
print(f"MSE     : {mse_GBR}")
print(f"R²      : {r2_GBR}")
print(f"RMSE    : {rmse_GBR}")
```

```python
df_eval.loc['GradientBoostingRegressor'] = [mae_GBR, mse_GBR, r2_GBR, rmse_GBR]
df_eval
```
